```python
from sklearn.cluster import KMeans
import numpy as np

X = np.array([[2, 10], [2, 5], [8, 4],[5, 8], [7, 5], [6, 4], [1, 2], [4, 9]])
kmeans = KMeans(n_clusters=3).fit(X) # 데이터가 필요한데 비지도 학습이니 y값이 없다

print("Labels: ", kmeans.labels_) # 전체 데이터 개수만큼 전체 데이터의 클러스터 결과
print("Cluster Centers: ", kmeans.cluster_centers_) # n 파라미터 개수의 센터를 지정
print("Predict Values: ", kmeans.predict([[1, 1]])) 
```

```python
import matplotlib. pyplot as plt

# 파라미터 : 좌료, 클러스터 결과
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_) # c가 클러스터 결과로 동일한 클러스터 별로 같은 색의 시각화를 보여준다
```

```python
# 파라미터 : 중심점 좌표
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color="black", marker = "+", s=200)
```

```python
import pandas as pd

df = pd.read_csv("r15.csv")

# 위에서의 X로 비지도인 X만 필요하다, 근데 타겟을 왜 구할까?? -> 평가지표에 넣기 위함이다(반드시 정수)
training_points = df[["col1", "col2"]]
training_labels = df["target"]

kmeans = KMeans(n_clusters = 15).fit(training_points)

plt.scatter(training_points["col1"], training_points["col2"], c=kmeans.labels_,cmap='rainbow')
```

```python
from sklearn.metrics.cluster import adjusted_rand_score

# 파라미터 : 실제 타겟, 예측타겟
arc = adjusted_rand_score(training_labels, kmeans.labels_)

print(arc)
```

```python
# spiral data
df = pd.read_csv("spiral.csv")
training_points = df[["col1", "col2"]]
training_labels = df["target"]

kmeans = KMeans(n_clusters = 3).fit(training_points)

plt.scatter(training_points["col1"], training_points["col2"], c=kmeans.labels_,cmap='rainbow')

arc = adjusted_rand_score(training_labels, kmeans.labels_)

print(arc)
```
