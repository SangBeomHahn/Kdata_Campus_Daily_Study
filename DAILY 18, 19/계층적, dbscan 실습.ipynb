```python
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
sample_df = pd.read_csv("r15.csv")
training_points = sample_df[["col1", "col2"]]
training_labels = sample_df["target"]
dbscan = DBSCAN(eps=0.6, min_samples=10).fit(training_points)
plt.scatter(training_points["col1"], training_points["col2"], c=dbscan.labels_,

cmap='rainbow')

plt.show()
```

```python
# 이정도 밀집도면 1.4에 10정도다

from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.cluster import adjusted_rand_score

sample_df = pd.read_csv("flame.csv")
training_points = sample_df[["col1", "col2"]]
training_labels = sample_df["target"]
eps_Me = 0
arr_Eps = []
arr_Acc = []

for i in range(1, 100):
  eps_Me += 0.01
  for j in range(1, 30):
    dbscan = DBSCAN(eps=eps_Me, min_samples=j).fit(training_points)
    acc = adjusted_rand_score(training_labels, dbscan.labels_)


    arr_Eps.append(eps_Me)
    arr_Acc.append(acc)

print(max(arr_Acc))
```
