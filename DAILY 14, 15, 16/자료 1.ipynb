```python
import numpy as np
import os
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.utils import np_utils # 여러가지 유틸을 사용하기 위한 유틸 다운
```

```python
np.random.seed(10)
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train = X_train.reshape(X_train.shape[0], 3072)
X_test = X_test.reshape(X_test.shape[0], 3072)
```

```python
# /255.0과 동일한 작업
X_train = (X_train - np.mean(X_train) / np.std(X_train))
X_test = (X_test - np.mean(X_test) / np.std(X_test))
```

```python
# 원 핫 벡터로 만들기 -> 이게 np.utils에 들어있는 거네
labels = 10

y_train = np_utils.to_categorical(y_train, labels)
y_test = np_utils.to_categorical(y_test, labels)
y_train
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense , Dropout
```

```python
input_layer = Input(shape=(3072,))
dense_layer_1 = Dense(512, activation="relu")(input_layer)
dropout_1 = Dropout(0.4)(dense_layer_1)
```

```python
dense_layer_2 = Dense(120, activation="relu")(dropout_1)
dropout_2 = Dropout(0.2)(dense_layer_2)

dense_layer_3 = Dense(labels, activation="sigmoid")(dropout_2)

model = Model(inputs=input_layer, outputs=dense_layer_3)
```

```python
model.summary()
```

```python
model.compile(loss = "categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, batch_size=1000, epochs = 10, validation_data=(X_test, y_test))
```

```python
loss, acc = model.evaluate(X_test, y_test)
```

```python
print("loss = ", loss, "acc = ", acc)
```

```python
#ㅋㅋㅋㅋㅋㅋㅋㅋ이거 왜 그러냐
model.predict(X_test).argmax(axis=1)
```

```python
model.save("model.h5")
```
